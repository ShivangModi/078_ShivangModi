```python
import nltk
from nltk.corpus import movie_reviews as mr
import matplotlib.pyplot as plt
import random
from collections import defaultdict
```

```python
reviews = nltk.download('movie_reviews')
```

```python
data = defaultdict(list)
for i in mr.fileids():
    data[i.split('/')[0]].append(i)

# print(data)
pos_review = data['pos'][:15]
print(pos_review)
neg_review = data['neg'][:10]
print(neg_review)
```

```python
fig = plt.figure(figsize=(5,5))
lab = '1+', 'Samsung', 'iPhone'
fav = [40, 50, 30]

plt.pie(fav, labels=lab, autopct='%.2f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.title('Pie Chart of Favourite Phone')
plt.show()
```

```python
print(f'Positive Reviews: {len(pos_review)}')
print(f'Negative Reviews: {len(neg_review)}\n')

print(f'The type of Positive Reviews: {type(pos_review)}')
print(f'The type of Negative Reviews: {type(neg_review[0])}')
```

```python
fig = plt.figure(figsize=(5,5))
lab = 've+ Review', 've- Review'
rev = [len(pos_review), len(neg_review)]

plt.pie(rev, labels=lab, autopct='%.2f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.title('Pie Chart of Movie Review')
plt.show()
```

```python
import string   # for string operations
from nltk.corpus import stopwords   # module for stop words that come with NLTK
from nltk.stem import PorterStemmer # module for stemming

nltk.download('stopwords')
```

```python
word = stopwords.words('english')
print(word)

all_word = [i.lower() for i in mr.words()]
 
# print first 15 words
print(all_word[:15])
all_word_without_stopword = [i for i in all_word if i not in word]
print(all_word_without_stopword[:15])
```

```python
import string

print(string.punctuation)
all_word_without_punctuation = [i for i in all_word if i not in string.punctuation]
print(all_word_without_punctuation[:15])
```

```python
print('\033[92m')
print(all_word_without_punctuation[:15])
print('\033[94m')

# Stemming class
stemmer = PorterStemmer() 

# Create an empty list to store the stems
review_stem = [] 
for i in all_word_without_punctuation[:15]:
    stem_word = stemmer.stem(i)  
    review_stem.append(stem_word)  
print(f'Stemmed Words: {review_stem}')
```
